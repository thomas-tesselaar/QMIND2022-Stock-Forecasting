```python
import pandas as pd
import numpy as np
```

```python
import random
```

```python
aapl = pd.read_csv("AAPL10Y.csv")
tsla = pd.read_csv("TSLA.csv")
gsptse = pd.read_csv("^GSPTSE10Y.csv")
```

```python
aapl = aapl[["Date", "Adj Close"]]
tsla = tsla[["Date", "Adj Close"]]
gsptse = gsptse[["Date", "Adj Close"]]
```

## AAPL

```python
aapl_return = aapl["Adj Close"][521]/aapl["Adj Close"][0]
aapl_return
```

```python
aapl["Buy and Hold"] = 100*aapl["Adj Close"]
```

```python
investment = [100*aapl["Adj Close"][0]]
for i in range(1,522):
    d = random.random()
    change = (aapl["Adj Close"][i]/aapl["Adj Close"][i-1])
    if change > 1:
        if d <= 0.7647:
            investment.append(investment[i-1]*change)
        else:
#             investment.append(investment[i-1]/change)
            investment.append(investment[i-1])
    else:
        if d <= 0.4074:
#             investment.append(investment[i-1]/change)
            investment.append(investment[i-1])
        else:
            investment.append(investment[i-1]*change)
        
                 
investment
```

```python
aapl["Strategy"] = pd.DataFrame(investment)
aapl
```

```python
aapl[["Buy and Hold", "Strategy"]].plot()
```

```python
aapl_return2 = aapl["Strategy"][521]/aapl["Strategy"][0]
aapl_return2
```
